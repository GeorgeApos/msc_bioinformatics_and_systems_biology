### Graded Assignment (week 2)
Normal brain functioning depends heavily on a constant availability of glucose. Glucose transport to the brain can be modelled following MM kinetics. 
During this week you have seen the full set of equations to describe the Michaelis-Menten (MM) kinetics:

\begin{aligned}
\frac{dS}{dt} &= k_r C - k_f ES  &\qquad (1a) \\
\frac{dE}{dt} &= (k_r + k_{cat}) C - k_f ES  &\qquad (1b)  \\
\frac{dC}{dt} &= k_f ES - (k_r + k_{cat}) C  &\qquad (1c) \\
\frac{dP}{dt} &= k_{cat} C  &\qquad (1d)
\end{aligned}

We are going to interpret the MM kinetics in terms of the glucose dynamics in the brain (see below Figure). The enzyme $E$ in the MM kinetics is assumed to represent the glucose transporter GLUT-1 in the blood-brain barrier (BBB), while the product $P$ represents the brain glucose. We can furthermore assume a constant cerebral metabolic glucose rate (indicated by CMR).

Eqs. (1a-d) can be rewritten as a reduced set of equations by making some assumptions. We assume that:

- The dynamic rates for enzyme binding are much higher than the other dynamics over a given time interval. As a result, the densities of $C$ and $E$ do not change, i.e., $\frac{dC}{dt}=\frac{dE}{dt}=0$. 
- Also, we consider  $E_T = E+C$.

This leads to the derivation of the following equation:
\begin{equation}
\frac{dP}{dt} = k_{cat} \frac{E_T S}{K_M + S} \qquad (2)
\end{equation}

Here $P$ is a product formed from substrate $S$, where the rate of change follows a saturation curve. In other words, there is a maximum because of limitations in the availability of enzymes and the time it takes to convert substrate to product.

### Question 2 (2 points):
Derive $\frac{dS}{dt}$ using the above assumptions, then show that $\frac{dS}{dt} = -\frac{dP}{dt}$.

<font color='red'>

With $E_T = E+C$ we get $C = E_T - E$. Then:

\begin{aligned}
\frac{dC}{dt} &= k_f ES - (k_r + k_{cat}) C = k_f ES - (k_r + k_{cat}) (E_T-E) \\
&= k_f ES - k_r E_T - k_{cat} E_T + k_r E + k_{cat}E = (k_f S + k_r + k_{cat})E - (k_r+k_{cat})E_T  = 0
\end{aligned}

\begin{equation}
(k_f S + k_r + k_{cat})E = (k_r+k_{cat})E_T
\end{equation}

\begin{equation}
E = \frac{(k_r+k_{cat})E_T}{k_f S + k_r + k_{cat}} = \frac{\frac{k_r+k_{cat}}{k_f}E_T}{S + \frac{k_r + k_{cat}}{k_f}} = \frac{K_M E_T}{S+K_M}
\end{equation}

(with $K_M = \frac{k_r+k_{cat}}{k_f}$)

Now substitute this into $\frac{dS}{dt}$:

\begin{equation}
\frac{dS}{dt} = k_r C - k_f ES = k_r (E_T - E) - k_f \frac{K_M E_T}{S+K_M} S = k_r \left(E_T - \frac{K_M E_T}{S+K_M}\right) - k_f \frac{K_M E_T}{S+K_M} S
\end{equation}

\begin{equation}
\frac{dS}{dt} = k_r E_T\frac{S+K_M}{S+K_M} - k_r\frac{K_M E_T}{S+K_M} - k_f S \frac{K_M E_T}{S+K_M} = \frac{k_r E_T S + k_r E_T K_M - k_r E_T K_M - k_f K_M E_T S}{S+K_M}
\end{equation}

Removing terms that cancel each other:

\begin{equation}
\frac{dS}{dt} = \frac{k_r E_T S - k_f K_M E_T S}{S+K_M} = \frac{(k_r - k_f K_M) E_T S}{S+K_M} = \frac{\left(k_r - k_f \frac{k_r+k_{cat}}{k_f} \right) E_T S}{S+K_M}
\end{equation}

\begin{equation}
\frac{dS}{dt} = -k_{cat} \frac{ E_T S}{S+K_M} = -\frac{dP}{dt}
\end{equation}

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

def ode_glucose(c,t,Gp,Tmax,Vd,Kt,CMR,Kb2):
    Gb = c[0]
    
    dGbdt = Tmax*(Gp-Gb/Vd)/(Kt+Gp+Gb/Vd) - CMR*(Gb/Kb2)/(1+Gb/Kb2)
    
    return dGbdt

t_steps = np.linspace(0,50,500)

Gp = 15
Tmax = 0.69
Vd = 0.77
Kt = 0.6
CMR = Tmax/4.5

Gb0 = 0

Gbs = []

Kb2 = 1
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_1 = x_t[:,0]

Kb2 = 2
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_2 = x_t[:,0]

Kb2 = 3
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_3 = x_t[:,0]

Kb2 = 4
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_4 = x_t[:,0]

Kb2 = 5
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_5 = x_t[:,0]

Kb2 = 6
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_6 = x_t[:,0]

Kb2 = 7
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_7 = x_t[:,0]

Kb2 = 8
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_8 = x_t[:,0]

Kb2 = 9
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_9 = x_t[:,0]

Kb2 = 10
x_t = odeint(ode_glucose, Gb0, t_steps, args = (Gp,Tmax,Vd,Kt,CMR,Kb2))
Gb_10 = x_t[:,0]


plt.plot(t_steps, Gb_1, label = '$K_{b2} = 1$')
plt.plot(t_steps, Gb_2, label = '$K_{b2} = 2$')
plt.plot(t_steps, Gb_3, label = '$K_{b2} = 3$')
plt.plot(t_steps, Gb_4, label = '$K_{b2} = 4$')
plt.plot(t_steps, Gb_5, label = '$K_{b2} = 5$')
plt.plot(t_steps, Gb_6, label = '$K_{b2} = 6$')
plt.plot(t_steps, Gb_7, label = '$K_{b2} = 7$')
plt.plot(t_steps, Gb_8, label = '$K_{b2} = 8$')
plt.plot(t_steps, Gb_9, label = '$K_{b2} = 9$')
plt.plot(t_steps, Gb_10, label = '$K_{b2} = 10$')

plt.legend()
plt.xlabel('$t$')
plt.ylabel('$G_{brain}$')

plt.show()
plt.close()
```
