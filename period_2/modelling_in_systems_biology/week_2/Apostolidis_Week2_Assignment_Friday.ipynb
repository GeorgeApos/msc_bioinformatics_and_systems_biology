### Graded Assignment (week 2)
Normal brain functioning depends heavily on a constant availability of glucose. Glucose transport to the brain can be modelled following MM kinetics.
During this week you have seen the full set of equations to describe the Michaelis-Menten (MM) kinetics:

\begin{aligned}
\frac{dS}{dt} &= k_r C - k_f ES  &\qquad (1a) \\
\frac{dE}{dt} &= (k_r + k_{cat}) C - k_f ES  &\qquad (1b)  \\
\frac{dC}{dt} &= k_f ES - (k_r + k_{cat}) C  &\qquad (1c) \\
\frac{dP}{dt} &= k_{cat} C  &\qquad (1d)
\end{aligned}

We are going to interpret the MM kinetics in terms of the glucose dynamics in the brain (see below Figure). The enzyme $E$ in the MM kinetics is assumed to represent the glucose transporter GLUT-1 in the blood-brain barrier (BBB), while the product $P$ represents the brain glucose. We can furthermore assume a constant cerebral metabolic glucose rate (indicated by CMR).

### Question (1 point):
Give the biological interpretation of the variable $S$ from Eqs. (1a-d) in the context of the above Figure.

***Solution:*** S is the substrate, here glucose on the plasma side of the blood–brain barrier (the capillary or extracellular glucose).

Eqs. (1a-d) can be rewritten as a reduced set of equations by making some assumptions. We assume that:

- The dynamic rates for enzyme binding are much higher than the other dynamics over a given time interval. As a result, the densities of $C$ and $E$ do not change, i.e., $\frac{dC}{dt}=\frac{dE}{dt}=0$.
- Also, we consider  $E_T = E+C$.

This leads to the derivation of the following equation:
\begin{equation}
\frac{dP}{dt} = k_{cat} \frac{E_T S}{K_M + S} \qquad (2)
\end{equation}

Here $P$ is a product formed from substrate $S$, where the rate of change follows a saturation curve. In other words, there is a maximum because of limitations in the availability of enzymes and the time it takes to convert substrate to product.

### Question 2 (2 points):
Derive $\frac{dS}{dt}$ using the above assumptions, then show that $\frac{dS}{dt} = -\frac{dP}{dt}$.

***Solution:***

Substitute \(E = E_T - C\) and solve for \(C\):

$$
0 = k_f (E_T - C) S - (k_r + k_{cat}) C
= k_f E_T S - C (k_f S + k_r + k_{cat}),
$$

$$
\Rightarrow\quad
C = \frac{k_f E_T S}{k_f S + k_r + k_{cat}}
= \frac{E_T S}{S + \frac{k_r + k_{cat}}{k_f}}
= \frac{E_T S}{S + K_M},
\qquad K_M = \frac{k_r + k_{cat}}{k_f}.
$$

Hence,

$$
\frac{dP}{dt} = k_{cat} C
= \frac{k_{cat} E_T S}{K_M + S}.
$$

From conservation of substrate units,

$$
\frac{d}{dt}(S + C + P) = 0
\quad\Rightarrow\quad
\frac{dS}{dt} = -\frac{dC}{dt} - \frac{dP}{dt},
$$

and under QSSA (\(dC/dt = 0\)) we get

$$
\boxed{\frac{dS}{dt} = -\frac{dP}{dt}}
\quad\Rightarrow\quad
\boxed{\frac{dS}{dt} = -\frac{k_{cat} E_T S}{K_M + S}}.
$$


Figure 1 does not fully correspond to Eqs. (1a-d). Choi et al. (2001) used a comparable model with *reversible* MM kinetics to make a quantitative match to *in vivo* data obtained using non-invasive nuclear magnetic resonance (NMR) methods on anesthetized rats. The difference between the ‘regular’ MM kinetics in Eqs. (1a-d) and reversible MM kinetics is that the product can again bind to the enzyme, and then possibly revert back to substrate again.

We now consider a reversible MM system, given as:
\begin{equation}
\begin{aligned}
E_2 + S_2 &\overset{k_{+1}}{\longrightarrow} C_2 \\
C_2 &\overset{k_{-1}}{\longrightarrow} E_2 + S_2 \\
C_2 &\overset{k_{+2}}{\longrightarrow} E_2 + P_2 \\
E_2 + P_2 &\overset{k_{-2}}{\longrightarrow} C_2
\end{aligned}
\end{equation}
This conceptually corresponds better to Figure 1.

### Question 3 (1 point):
Derive the ODE equations based on the above chemical reactions (Hint: there are again four equations, namely for $S_2,E_2,C_2$ and $P_2$).


***Solution:*** The corresponding mass–action ODE system is:
$$
\begin{aligned}
\frac{dS_2}{dt} &= -\,k_{+1}\,E_2 S_2 \;+\; k_{-1}\,C_2,\\[4pt]
\frac{dP_2}{dt} &= \phantom{-}\,k_{+2}\,C_2 \;-\; k_{-2}\,E_2 P_2,\\[6pt]
\frac{dC_2}{dt} &= \phantom{-}\,k_{+1}\,E_2 S_2 \;+\; k_{-2}\,E_2 P_2 \;-\; (k_{-1}+k_{+2})\,C_2,\\[6pt]
\frac{dE_2}{dt} &= -\,k_{+1}\,E_2 S_2 \;-\; k_{-2}\,E_2 P_2 \;+\; (k_{-1}+k_{+2})\,C_2.
\end{aligned}
$$


### Question 4 (2 points):
Use similar assumptions as before to show that $\frac{dP_2}{dt} = f\left( \frac{S_2}{K_{M1}}, \frac{P_2}{K_{M2}} \right)$, where $K_{M1} = \frac{k_{-1}+k_{+2}}{k_{+1}}$ and $K_{M2} = \frac{k_{-1}+k_{+2}}{k_{-2}}$.

**Solution:Using QSSA on \(C_2\) and \(E_T = E_2 + C_2\):**

$$
\frac{dC_2}{dt}
= k_{+1}E_2S_2 + k_{-2}E_2P_2 - (k_{-1}+k_{+2})C_2 = 0,
\qquad
E_2 = E_T - C_2.
$$

**Solve for \(C_2\):**

$$
0 = k_{+1}(E_T - C_2)S_2 + k_{-2}(E_T - C_2)P_2 - (k_{-1}+k_{+2})C_2
$$

$$
\Rightarrow\quad
C_2
= \frac{E_T\,(k_{+1}S_2 + k_{-2}P_2)}
{k_{+1}S_2 + k_{-2}P_2 + (k_{-1}+k_{+2})},
\qquad
E_2
= \frac{E_T\,(k_{-1}+k_{+2})}
{k_{+1}S_2 + k_{-2}P_2 + (k_{-1}+k_{+2})}.
$$

**Net product rate:**

$$
\frac{dP_2}{dt}
= k_{+2}C_2 - k_{-2}E_2P_2
= \frac{E_T\big(k_{+2}k_{+1}S_2 - k_{-2}k_{-1}P_2\big)}
{k_{+1}S_2 + k_{-2}P_2 + (k_{-1}+k_{+2})}.
$$

**Define the reversible MM constants and simplify:**

$$
K_{M1} = \frac{k_{-1}+k_{+2}}{k_{+1}},
\qquad
K_{M2} = \frac{k_{-1}+k_{+2}}{k_{-2}},
\qquad
K_{M3} = k_{+2}E_T.
$$

Dividing numerator and denominator by \((k_{-1}+k_{+2})\) gives

$$
\boxed{
\frac{dP_2}{dt}
=
\frac{K_{M3}\left(\dfrac{S_2}{K_{M1}} - \dfrac{P_2}{K_{M2}}\right)}
{1 + \dfrac{S_2}{K_{M1}} + \dfrac{P_2}{K_{M2}}}
}
$$

which is a function of \(S_2/K_{M1}\) and \(P_2/K_{M2}\), as required.

### Question 7 (1 point):
Give the units of all the $K.$ parameters, where the dot index stands for different indices (such as $M1$).

Assume that time $t$ is in minutes (min) and concentrations are in $\mu\text{mol}/\text{g}$.  
Then $G_B$ is in $\mu\text{mol}/\text{g}$; $k_{+1}$ and $k_{-2}$ have units $\text{g}/(\mu\text{mol}\cdot\text{min})$;  
$k_{-1}$ and $k_{+2}$ have units $1/\text{min}$._

Now assume the following model for the glucose concentration in the brain, now written as $G_{brain}$:

\begin{equation}
\frac{dG_{brain}}{dt} = T_{max} \frac{G_{plasma} - G_{brain}/V_d}{K_t + G_{plasma} + G_{brain}/V_d} - CMR \frac{G_{brain}/K_{b2}}{1+G_{brain}/K_{b2}}
\end{equation}

Based on the data by Choi et al. (2001), we have the following parameter values:

$T_{max} = 0.69 \mu mol/(g \cdot min)$, $K_t = 0.6 mM$, $G_{plasma} = 4.6-29 mM$, and $V_d = 0.77 ml/g$.

Furthermore, $\frac{T_{max}}{CMR} = 4.5$.

### Question 8 (1 point):
Code Eq. (3) into Jupyter, and simulate the model from 0 to 50 minutes, assuming that initially the glucose in the brain is zero. Increase parameter $K_{b2}$ from 1 to 10 in equidistant steps of 1, and provide some conclusions about what you see in the simulations. Select any value for $G_{plasma}$ and keep it fixed across the simulations.

```python
# Brain glucose model (Choi et al. 2001-inspired)
# dG_B/dt = Tmax * (Gplasma - G_B/Vd) / (Kt + Gplasma + G_B/Vd)  -  CMR * G_B / (Kb2 + G_B)
# Units (as typically reported): Tmax [umol/(g*min)], Kt [mM], Gplasma [mM], Vd [ml/g]
# "TmaxCMR = 4.5" is interpreted as Tmax / CMR = 4.5  ->  CMR = Tmax / 4.5

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# -------- parameters --------
Tmax = 0.69      # umol/(g*min)
Kt   = 0.6       # mM
Vd   = 0.77      # ml/g
Gplasma = 5.0    # mM (choose any fixed value in [4.6, 29])
CMR = Tmax / 4.5 # umol/(g*min)  (from "Tmax/CMR = 4.5")

# -------- time span and initial condition --------
t_span = (0.0, 50.0)             # minutes
t_eval = np.linspace(*t_span, 600)
G0 = [0.0]                       # initial brain glucose (umol/g)

# -------- simulation for Kb2 = 1..10 --------
plt.figure(figsize=(7,5))
final_values = []

for Kb2 in range(1, 11):         # 1,2,...,10 mM
    def rhs(t, G, Kb2=Kb2):
        Gb = float(G)
        influx = Tmax * ((Gplasma - Gb / Vd) / (Kt + Gplasma + Gb / Vd))
        consumption = CMR * (Gb / (Kb2 + Gb))
        return influx - consumption

    sol = solve_ivp(rhs, t_span, G0, t_eval=t_eval, rtol=1e-8, atol=1e-10)
    plt.plot(sol.t, sol.y[0], label=f"Kb2={Kb2} mM")
    final_values.append((Kb2, sol.y[0, -1]))

plt.title("Brain glucose dynamics for varying $K_{b2}$")
plt.xlabel("Time (min)")
plt.ylabel(r"$G_B$ (µmol/g)")
plt.grid(True)
plt.legend(ncol=2, fontsize=8)
plt.show()

# Print the end-of-simulation values as steady-state approximations
for kb, val in final_values:
    print(f"Kb2={kb:2d} mM  ->  G_B(50 min) ≈ {val:.4f} µmol/g")
```
